```python
print(dados["Quantas horas você costuma dormir por dia?"].unique())
```

**Reasoning**:
Create a mapping from the text categories to representative numerical values based on the identified unique categories.



```python
sleep_hours_mapping = {
    'Menos de 5 horas': 4,
    'Entre 6 a 9 horas': 7.5,
    'Mais de 10 horas': 11
}
```

```python
dados["Quantas horas você costuma dormir por dia?"] = dados["Quantas horas você costuma dormir por dia?"].map(sleep_hours_mapping)
```
